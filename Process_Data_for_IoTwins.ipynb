```python
import pandas as pd

out_folder = './servises/super_semisuper_anomaly_detection_services/data'
csv_file_name = 'caravan-insurance-challenge'
ext = '.csv'
```

```python
# Read in data and separate into training and testing sets
data = pd.read_csv('data-example/'+csv_file_name+ext)
train = data[data['ORIGIN'] == 'train']
test = data[data['ORIGIN'] == 'test']

# Drop the unneeded columns
train = train.drop(columns = ['ORIGIN'])
test = test.drop(columns = ['ORIGIN'])

# Rename the last column to 'label'
train.rename(columns={'CARAVAN': 'label'}, inplace=True)
test.rename(columns={'CARAVAN': 'label'}, inplace=True)
```
